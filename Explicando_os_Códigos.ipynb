Linguagem Booleana

Criado por George Boole em 1847

Sistema matemático que trata variáveis lógicas, assumindo apenas dois valores:
0 (false/desligado) 1 (true/ligado)

Em Python, consideramos os valores booleanos como True e False

```python
texto = "Falso"   #<--- dentro dos aspas caracteriza-se como texto, se a aspas tiver vazia resultado vai ser falso
if texto:
  print("Verdadeiro")
else:
  print("Falso")
```

```python
num = 0     #<--- sem aspas caracteriza-se como número, se for sem nada sera falso.
if num:
  print("Verdadeiro")
else:
  print("Falso")
```

```python
#Crie um código em Python que solicite ao usuário dois valores numéricos (v1 e v2)

#Solicite também a operação a ser realizada (soma, subtração, multiplicação, divisão, potencial, módulo)

#Realize a operação selecionada pelo usuário e apresenteo resultado na tela

#Trate os possíveis problemas que podem ocorrer com as entradas numéricas.


#/////////////////////////////////////////////////////
#//                  Código                        //
#////////////////////////////////////////////////////

V1 = input("Coloque o valor 1: ") #<---- Entrada de valores
V2 = input("Coloque o valor 2: ") #<---- Entrada de valores

# verifica se os valores são números
if V1.replace('.', '', 1).isdigit() and V2.replace('.', '', 1).isdigit():

    V1 = float(V1) #<---- transforma em números
    V2 = float(V2) #<---- transforma em números

    print("\nEscolha a operação:") #<---- Print para instrução
    print("soma") #<---- print do tipo de operação
    print("subtracao") #<---- print do tipo de operação
    print("multiplicacao") #<---- print do tipo de operação
    print("divisao") #<---- print do tipo de operação
    print("potencia") #<---- print do tipo de operação
    print("modulo") #<---- print do tipo de operação

    operacao = input("Digite a operação: ") #<------ Input do tipo de operação a cima

    if operacao == "soma": #<---- If da operação selecionada.
        print("Resultado:", V1 + V2)

    elif operacao == "subtracao": #<---- elif da operação selecionada.
        print("Resultado:", V1 - V2)

    elif operacao == "multiplicacao": #<---- elif da operação selecionada.
        print("Resultado:", V1 * V2)

    elif operacao == "divisao": #<---- elif da operação selecionada.
        if V2 == 0:
            print("Erro: divisão por zero")
        else:
            print("Resultado:", V1 / V2)

    elif operacao == "potencia": #<---- elif da operação selecionada.
        print("Resultado:", V1 ** V2)

    elif operacao == "modulo": #<---- elif da operação selecionada.
        print("Resultado:", V1 % V2)

    else:
        print("Operação inválida")

else: #<---- Se o usuário digitar valores não númericos ou strings de texto.
    print("Erro: digite apenas números")


# O elif após o if cai em verificação de item por item até chegar na opção selecionada pelo usuário
# ele cria um tipo de verificação linha a linha.
```

```python
import tkinter as tk  # <--- importa o módulo tkinter, usado para criar interfaces gráficas. O "as tk" serve para abreviar o nome e facilitar a escrita.

# função para adicionar números e operadores no visor
def clicar(valor):  # <--- "def" é usada para definir uma função. Essa função recebe um valor e coloca no visor.
    entrada_atual = visor.get()  # <--- pega o conteúdo que já está escrito no visor.
    visor.delete(0, tk.END)  # <--- apaga tudo do visor, do início (0) até o final (tk.END).
    visor.insert(0, entrada_atual + str(valor))  # <--- insere no visor o conteúdo anterior mais o novo valor clicado.

# função para limpar o visor
def limpar():  # <--- cria a função que vai limpar tudo que estiver no visor.
    visor.delete(0, tk.END)  # <--- apaga todo o conteúdo do visor.

# função para calcular o resultado
def calcular():  # <--- cria a função responsável por fazer o cálculo da expressão digitada.
    try:  # <--- tenta executar o código abaixo. Serve para evitar que o programa quebre em caso de erro.
        conta = visor.get()  # <--- pega o que foi digitado no visor e guarda na variável "conta".
        resultado = eval(conta)  # <--- "eval" calcula a expressão matemática que está em forma de texto.
        visor.delete(0, tk.END)  # <--- limpa o visor antes de mostrar o resultado.
        visor.insert(0, str(resultado))  # <--- mostra o resultado no visor, convertido para texto.
    except:  # <--- se acontecer algum erro no cálculo, executa o bloco abaixo.
        visor.delete(0, tk.END)  # <--- limpa o visor.
        visor.insert(0, "Erro")  # <--- mostra a palavra "Erro" no visor.

# criar janela principal
janela = tk.Tk()  # <--- cria a janela principal da aplicação.
janela.title("Calculadora Edypho Bandeira")  # <--- define o título que aparecerá na barra superior da janela.
janela.geometry("450x600")  # <--- define o tamanho da janela: 450 de largura por 600 de altura.
janela.resizable(False, False)  # <--- impede que a janela seja redimensionada, tanto na largura quanto na altura.
janela.configure(bg="lightgray")  # <--- define a cor de fundo da janela como cinza claro.

# visor da calculadora
visor = tk.Entry(janela, font=("Arial", 24), bd=10, relief="sunken", justify="right")
# <--- cria o campo de entrada (visor) dentro da janela.
# "font" define a fonte e o tamanho do texto.
# "bd=10" define a espessura da borda.
# "relief='sunken'" deixa o visor com aparência afundada.
# "justify='right'" alinha o texto à direita.

visor.grid(row=0, column=0, columnspan=4, padx=10, pady=20, ipadx=8, ipady=15)
# <--- posiciona o visor na linha 0 e coluna 0 da grade.
# "columnspan=4" faz o visor ocupar 4 colunas.
# "padx" e "pady" criam espaço por fora.
# "ipadx" e "ipady" aumentam o espaço interno.

# botões
botoes = [
    ('7', 1, 0), ('8', 1, 1), ('9', 1, 2), ('/', 1, 3),  # <--- primeira linha de botões
    ('4', 2, 0), ('5', 2, 1), ('6', 2, 2), ('*', 2, 3),  # <--- segunda linha de botões
    ('1', 3, 0), ('2', 3, 1), ('3', 3, 2), ('-', 3, 3),  # <--- terceira linha de botões
    ('0', 4, 0), ('.', 4, 1), ('=', 4, 2), ('+', 4, 3),  # <--- quarta linha de botões
    ('C', 5, 0)  # <--- quinta linha com o botão de limpar
]

# criar os botões na tela
for (texto, linha, coluna) in botoes:  # <--- percorre cada item da lista "botoes", separando o texto, a linha e a coluna.
    if texto == "=":  # <--- verifica se o botão atual é o sinal de igual.
        botao = tk.Button(janela, text=texto, width=8, height=3, font=("Arial", 14), command=calcular)
        # <--- cria o botão "=".
        # "text=texto" define o texto que aparece no botão.
        # "width" e "height" definem o tamanho.
        # "font" define a fonte.
        # "command=calcular" faz com que ao clicar chame a função calcular.

    elif texto == "C":  # <--- verifica se o botão atual é o botão de limpar.
        botao = tk.Button(janela, text=texto, width=35, height=3, font=("Arial", 14), command=limpar)
        # <--- cria o botão "C" com largura maior.
        # "command=limpar" chama a função limpar ao clicar.

        botao.grid(row=linha, column=coluna, columnspan=4, padx=5, pady=5)
        # <--- posiciona o botão "C" na grade.
        # "columnspan=4" faz ele ocupar 4 colunas inteiras.
        continue  # <--- faz o laço pular para o próximo item, sem executar o restante abaixo.

    else:  # <--- para todos os outros botões, como números e operadores.
        botao = tk.Button(
            janela,
            text=texto,
            width=8,
            height=3,
            font=("Arial", 14),
            command=lambda t=texto: clicar(t)
        )
        # <--- cria os botões normais.
        # "lambda t=texto: clicar(t)" serve para passar o valor do botão clicado para a função clicar.

    botao.grid(row=linha, column=coluna, padx=5, pady=5)
    # <--- posiciona o botão na linha e coluna correspondentes da grade.
    # "padx" e "pady" criam espaçamento entre os botões.

# iniciar a calculadora
janela.mainloop()  # <--- inicia o loop principal da janela, fazendo a interface aparecer e funcionar.

#pra executar e formar a calculadora na tela, tem que salvar ele como:
# calculadora.py e executar ele no prompt de comando do software que usa para codar ex: vscode e pycharm.
#Linha de execução após salvar: python calculadora.py depende do local que você salvou, caso tenha salvo em algum lugar especifique o caminho
# igual o exemplo a seguir "python Documents\Aulas\Calculadora.py"
```

Laços de Repetição

Também chamados de loops.

while - Repete o código baseado em uma condição booleana, ele não sabe quantas vezes irá repetir, Alto risco de loop infinito, repete "enquanto|" uma condição for verdadeira


for- Repete sobre uma coleção ou intervalos fixos, geralmente sabe quantas vezes vai repetir, baixo risco de loop infinito, repete "para" cada item.

```python
# Sistema de pedidos de uma pastelaria

# O programa deve permitir que o usuário monte seu pedido de pastéis.

# O usuário deve escolher:
# - Tipo de pastel: doce ou salgado
# - Tamanho do pastel: pequeno, médio ou grande

# Cada combinação de tipo e tamanho possui um preço diferente.

# O sistema deve permitir que o usuário faça vários pedidos,
# ou seja, adicionar mais de um pastel.

# O programa deve calcular o valor total da compra conforme os pedidos são feitos.

# Ao final, o usuário deve ter a opção de encerrar o pedido.

# Quando o pedido for encerrado, o sistema deve mostrar:
# - O total a pagar

# O programa deve utilizar:
# - Estruturas condicionais (if, elif, else)
# - Estruturas de repetição (while)
# - Entrada de dados com input()
# - Saída de dados com print()
# - Operações matemáticas para calcular o total

#//////////////////////////////////////////////////////////////////////////////
#/////                          Lição de Casa                             /////
#//////////////////////////////////////////////////////////////////////////////




total = 0  # variável que guarda o total do pedido

while True:  # loop infinito (fica repetindo até usar break)

    print("\n--- PASTELARIA ---")  # mostra o título
    print("1 - Doce")  # opção 1
    print("2 - Salgado")  # opção 2
    print("0 - Fechar pedido")  # opção para sair

    opcao = input("Escolha o tipo: ")  # usuário digita a opção

    if opcao == "0":  # se escolher 0
        break  # sai do loop (encerra o programa)

    if opcao == "1":  # se escolher doce
        sabor = "Doce"  # define sabor como doce

    elif opcao == "2":  # se escolher salgado
        sabor = "Salgado"  # define sabor como salgado

    else:  # se digitar algo diferente
        print("Opção inválida!")  # mostra erro
        continue  # volta pro início do loop

    print("\nTamanhos:")  # título da escolha de tamanho
    print("1 - Pequeno")  # opção pequeno
    print("2 - Médio")  # opção médio
    print("3 - Grande")  # opção grande

    tamanho_opcao = input("Escolha o tamanho: ")  # usuário escolhe tamanho

    if tamanho_opcao == "1":  # se escolher pequeno
        tamanho = "P"  # define tamanho P

        if sabor == "Doce":  # se for doce
            preco = 5  # preço do doce pequeno
        else:  # se for salgado
            preco = 6  # preço do salgado pequeno

    elif tamanho_opcao == "2":  # se escolher médio
        tamanho = "M"  # define tamanho M

        if sabor == "Doce":  # se for doce
            preco = 7  # preço do doce médio
        else:  # se for salgado
            preco = 8  # preço do salgado médio

    elif tamanho_opcao == "3":  # se escolher grande
        tamanho = "G"  # define tamanho G

        if sabor == "Doce":  # se for doce
            preco = 9  # preço do doce grande
        else:  # se for salgado
            preco = 10  # preço do salgado grande

    else:  # se digitar tamanho inválido
        print("Tamanho inválido!")  # mostra erro
        continue  # volta pro início do loop

    quantidade = int(input("Quantos pastéis deseja? "))  # pede quantidade e transforma em número

    subtotal = preco * quantidade  # calcula o valor desse pedido
    total = total + subtotal  # soma no total geral

    print(f"\nVocê pediu {quantidade} pastel(is) {sabor} tamanho {tamanho}")  # mostra o pedido
    print(f"Subtotal: R${subtotal}")  # mostra subtotal
    print(f"Total até agora: R${total}")  # mostra total acumulado

print("\n--- PEDIDO FINALIZADO ---")  # mensagem final
print(f"Total a pagar: R${total}")  # mostra total final

input("Pressione ENTER para sair...")  # pausa o programa
```

As listas possuem onze operações possiveis

appent(item) - Adiciona um item ao final da lista.
clear() - Limpa a lista.
count(item) - Retorna o número de ocorrências de item na lista.
copy( ) - Retorna uma cópia da lista.
index(item) - Retorna o índice da primeira ocorrência do item.
insert(idx, item) - Insere item no index idx.
pop( ) - Remove o último item da lista.
remove(item) - Remove a primeira ocorrência de item na lista.
reverse( ) - Inverte a ordem da lista.
extend(L) - Insere cada elemento da lista L dentro da lista atual.
